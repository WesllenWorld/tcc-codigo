```python
#trocar muški por 0 e ženski por 1 em todas as linhas da coluna 50 usando iloc
#muski significa masculino e zenski significa feminino

df.iloc[:,50] = df.iloc[:,50].replace('muški', 0)
df.iloc[:,50] = df.iloc[:,50].replace('ženski', 1)



#trocar Srednja škola por 1, Osnovna škola por 0
#Srednja škola significa ensino médio e Osnovna škola significa ensino fundamental
df.iloc[:,51] = df.iloc[:,51].replace('Srednja škola', 1)
df.iloc[:,51] = df.iloc[:,51].replace('Osnovna škola', 0)


#trocar Veoma loša por 0, Skladu sa mojim potrebama por 1, Veoma dobra por 2
#"Veoma loša" seria traduzido como "Muito ruim", "Skladu sa mojim potrebama" como "De acordo com minhas necessidades", e "Veoma dobra" como "Muito boa".
df.iloc[:,52] = df.iloc[:,52].replace('Veoma loša', 0)
df.iloc[:,52] = df.iloc[:,52].replace('U skladu sa mojim potrebama', 1)
df.iloc[:,52] = df.iloc[:,52].replace('Veoma dobra', 2)


#trocar Nema por 0, osnovna škola por 1, srednja škola por 2, fakultet por 3
#Nema significa "Não tem", osnovna škola significa "Ensino fundamental", srednja škola significa "Ensino médio" e fakultet significa "Faculdade"
df.iloc[:,53] = df.iloc[:,53].replace('Nema', 0)
df.iloc[:,53] = df.iloc[:,53].replace('Osnovna škola', 1)
df.iloc[:,53] = df.iloc[:,53].replace('Srednja škola', 2)
df.iloc[:,53] = df.iloc[:,53].replace('Fakultet', 3)

#trocar Nema por 0, osnovna škola por 1, srednja škola por 2, fakultet por 3
#Nema significa "Não tem", osnovna škola significa "Ensino fundamental", srednja škola significa "Ensino médio" e fakultet significa "Faculdade"
df.iloc[:,54] = df.iloc[:,54].replace('Nema', 0)
df.iloc[:,54] = df.iloc[:,54].replace('Osnovna škola', 1)
df.iloc[:,54] = df.iloc[:,54].replace('Srednja škola', 2)
df.iloc[:,54] = df.iloc[:,54].replace('Fakultet', 3)

df.head(100)

```

```python
#selecionar as variaveis categoricas do df
#e fazer um one hot encoded usando get_dummies
d_situacao_financeira = pd.get_dummies(df.iloc[:,52], prefix='situação financeira', dtype=int)
d_escolaridade_materna = pd.get_dummies(df.iloc[:,53], prefix='escolaridade materna', dtype=int)
d_escolaridade_paterna = pd.get_dummies(df.iloc[:,54], prefix='escolaridade paterna', dtype=int)

columns = ['Situação financeira: Muito ruim', 'Situação financeira: De acordo com minhas necessidades', 'Situação financeira: Muito boa']
d_situacao_financeira.columns = columns

columns = ['Escolaridade da mãe: Não tem', 'Escolaridade da mãe: Ensino fundamental', 'Escolaridade da mãe: Ensino médio', 'Escolaridade da mãe: Faculdade']
d_escolaridade_materna.columns = columns

columns = ['Escolaridade do pai: Não tem', 'Escolaridade do pai: Ensino fundamental', 'Escolaridade do pai: Ensino médio', 'Escolaridade do pai: Faculdade']
d_escolaridade_paterna.columns = columns
d_escolaridade_materna

df= df.drop(df.columns[52:55], axis=1)

```

```python
#clusters para o k means
k = 2

from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
kmeans = KMeans(n_clusters=k)
clusters = kmeans.fit_predict(df)



rotulos = pd.DataFrame(clusters)

#usar o matplotlib para printar rotulos

plt.scatter(df.iloc[:,0], df.iloc[:,1], c=clusters, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')

plt.show()

```

## com standard


```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scalling = StandardScaler()

#fit e transform podem ser colocados em um comando só, como visto no exemplo do PCA
df_scaled = scalling.fit(df)
df_scaled = scalling.transform(df)

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

k = 2
kmeans = KMeans(k)
clusters = kmeans.fit_predict(df_pca)
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=clusters, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Clusters após PCA e KMeans')
plt.colorbar(label='Cluster')
plt.show()
```

```python
#aplicar silhouette score
from sklearn.metrics import silhouette_score
silhouette_score(df_pca, clusters)
```

```python
#aplicar davies bouldin score
from sklearn.metrics import davies_bouldin_score

davies_bouldin_score(df_pca, clusters)
```
